# Faster R-CNN

Faster R-CNNはマイクロソフト社が開発した世界初のDeep LearningによるEnd to End学習の物体検出モデル。  
A:物体か背景か学習し物体の場所を推定する工程 ①②  
B:物体が何か推定するクラス分類 (①)③④  
ABを交互に行う。  

Faster R-CNNは  
①入力画像から特徴マップを出力する工程  
②物体が写っている領域を推定するReagion Proposal Network  
③特徴マップを固定長に変換するReagion of interest  
④ROI後の特徴マップをクラス分類する工程  

①入力画像か ら特徴マップを出力する工程  
input       224,224,  3  
conv 3x3    224,224, 64   
conv 3x3    224,224, 64  
max pooling 112,112,128  
conv 3x3    112,112,128  
conv 3x3    112,112,128  
max pooling  56, 56,256  
conv 3x3     56, 56,256  
conv 3x3     56, 56,256  
conv 3x3     56, 56,256  
max pooling  28, 28,512  
conv 3x3     28, 28,512  
conv 3x3     28, 28,512  
conv 3x3     28, 28,512  
max pooling  14, 14,512  
conv 3x3     14, 14,512  
conv 3x3     14, 14,512  
conv 3x3     14, 14,512  
②物体が写っている領域を推定するReagion Proposal Network  
②-1 特徴マップのピクセル中心にAnchorを配置、Anchorを基準に複数のAnchor Boxを生成する。  
　　　Anchor Boxは基準の長さ×縦横比の複数生成する。Anchor Boxの面積は揃える。  
     k = 3 (length )x 3 (height width ratio)  
     Anchor Box = 14 x 14 (feature map)   
②-2 各Anchorについて  
　　　背景か物体か  
     ground truthとAnker boxのIntersection over Unionを計算し、  
     IoU < 0.3 なら背景  
     IoU > 0.7 なら物体  
     0.3 < IoU < 0.7 は使用しない。  
     class 2(object or not) * k (Anchor box)  
     どの程度ずれているか 4k(中心のズレx,y)*（長さのズレx,y）  
     conv 1x1 14, 14,2k と 14, 14,4k  
     RPNの出力は 14 x 14 x 2 x k に バイナリクロスエントロピー 2値問題  
     　　　　　　 14 x 14 x 4 x k に L1ノルム 回帰問題  
③ Faster R-CNN 物体のクラス分類  
③-1 ①の特徴マプを取得  
③-2 ROI Poolingを使用し特徴マップを固定長サイズにする。  
③-3 4096の全結合層に2回通し最後にクラス分類用と矩形ズレ回帰用の2種類を得る。  

Faster R-CNNの前身R-CNN、Fast R-CNNではPRNの代わりにSelective Searchが用いられている。  
Selective Searchは物体候補領域を切り出すアルゴリズムで高負荷な処理が必要。
