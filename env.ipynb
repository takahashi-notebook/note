## シェル
シェルとはOSのカーネルを操作するソフトウェア。  
シェルの入出力を表示するターミナルを用いて操作する。  
シェルの種類により、機能が異なるが、環境構築では標準のもので十分。  

階層構造のディレクトリを作成し、  
好きな場所にファイルを格納したり、プログラミングをインストールしたりする。  
基本的な操作は  
・作業するディレクトリ場所の確認  
・格納されているリストの確認  
・ファイルの確認、実行、移動、削除
