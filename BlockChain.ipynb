# ブロックチェーン

## ブロックチェーンとは
トランザクションの履歴（台帳）をブロックにまとめて、前回のブロックのハッシュを組み込みつつ、チェーンのように時系列につなげることで、一部の履歴を改竄すると組み込まれたハッシュ値がと整合性がなくなることを利用し、履歴が改竄されていないことを保証する技術。  
ブロックチェーン形成プロトコル（アルゴリズム）とコンセンサスプロトコル（アルゴリズム）に大別できる。

## トランザクションの暗号技術
暗号技術によりトランザクションの匿名性を担保できる。ビットコインでは楕円曲線暗号技術を利用した公開鍵とデジタル署名により トランザクションの匿名性と真正性を担保する。  

## ブロックチェーンの形成アルゴリズム
ブロックチェーンはコンセンスアルゴリズムに応じて形成される。  

ビットコインではコンセンサスアルゴリズムとして、必要な計算が成功することでキュリティを証明するプルーフオブワーク方式が用いられている。具体的には、前回のブロックチェーンのサマリーのハッシュの先頭が所定数連続して続くNonceを見つかると新しいブロックが形成され、見つけたノードに報奨が与えられる。P2Pの動機によるタイムラグで異なるブロックが形成された場合は、一番長いブロックチェーンを採用する。

参加者が変化してもブロックチェーンの生成平均時間が変化しないようにナンスが見つかる難易度を変更する。

手数料により優先順位が変化する。
 
IOTAではブロックを作らずにトランザクションをつなげる有向非巡回グラフ方式を用いている。  
 
パブリックのブロックチェーンでは、P2P技術を利用したブロックチェーンの自動生成システムとブロックチェーン作成の報酬により、システムに参加しているコンピュータがインセンティブを求めて競うことで、システムを自律的に維持しつつ、ブロックチェーンの真正性を保証する。  
 
プライベートのブロックチェーンでは、ブロック生成時の難易度やインセンティブが下げられ、トランザクション速度を上げている。  

通常のフォークは一時的ものであり、コンセスサスアルゴリズムにより採用されるブロックチェーンが決定される。  
アルゴリズム変更により別れる分岐をハードフィオークと言い、イーサでは新旧の2つある。


## ブロックチェーンの種類  
ビットコイン(block chain 1.0)  
トランザクションが非チューリング完全なブロックチェーン  
  
イーサリアム(block chain 2.0)  
トランザクションがチューリング完全なブロックチェーン  

FT 量的に換算でき交換可能なトークンを扱うブロックチェーン。通貨などに用いる。イーサでの規格ERC-20を参照
NFT 量的に換算できない交換不可能なトークンを扱うブロックチェーン。所有の証明などに用いられる。イーサでの規格ERC-721を参照

## 楕円曲線暗号理論  
有理数上の楕円曲線の積和演算は容易だが、剰余演算の難易度が高いために、公開鍵方式で利用されている暗号技術。ビットコインではsecp256k1が使用されている。

## 楕円曲線暗号理論の実装手順
1. 有限体での演算を実装する
2. 有限体上での楕円曲線の演算を実装する
3. 有限体上での楕円曲線のパラメータを決定する(secp256k1)

## 有限体の実装

インスタンスの演算が有限体での演算になるにクラスを実装する。 

'''
finite field is finite set has
additive identity, additive inverse,
multiplicative identity and multiplicative inverse.
finite field has p elements is
Fp = {0, 1, 2, .... , p-1 }
when p is product of prime numbers
in finite field when p is prime number
add in field = add & mod p
sub in field = sub & mod p
mul in field = mul & mod p
div in field = num ** (p-2) & mod p

Fields Elementnum class when num is element of Fp 
p is product of prime number
num in range(p)
'''
